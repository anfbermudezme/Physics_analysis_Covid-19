```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np 
```

```python
url = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv'
df = pd.read_csv(url,index_col=0,parse_dates=[0])
```

```python
df.head(10)
```

```python
world_table = df.loc[df.loc[:,'iso_code']=='OWID_WRL']
```

```python
df.loc[df.loc[:,'iso_code']=='OWID_WRL']
```

```python
data = {'date': list(world_table['date']),
        'total_vaccinations': list(world_table['total_vaccinations']),
        'people_vaccinated': list(world_table['people_vaccinated']),
        'people_fully_vaccinated': list(world_table['people_fully_vaccinated'])}
```

```python
DataFrame = pd.DataFrame(
    data,
    columns=[
        'date',
        'total_vaccinations',
        'people_vaccinated',
        'people_fully_vaccinated',
    ]
)
```

```python
DataFrame
```

```python
DataFrame.to_excel(r'world_vaccination.xlsx', index = False)
```

```python
DeathsVaccination = pd.read_excel('WorldVaccinationDeathsRelation.xlsx')
```

```python
x = np.arange(0, 562, 1)
y1 = DeathsVaccination['total_vaccinations']
y2 = DeathsVaccination['people_vaccinated']
y3 = DeathsVaccination['people_fully_vaccinated']
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1,'r--', label = "total Vaccinations")
ax.plot(x, y2,'k--', label = "people Vaccinated")
ax.plot(x, y3,'g--', label = "people Fully Vaccinated")
ax.legend(loc="upper left")
ax.set_xlabel("Date")
ax.set_ylabel("Agents")
plt.grid()
```
